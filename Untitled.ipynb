```python
import pandas as pd
```

```python
df = pd.read_csv('Earthquakes_database.csv')
```

```python
df.head()
```

```python
df['Date'] = pd.to_datetime(df['Date'], format='%m/%d/%Y', errors='coerce')
df['Year'] = df['Date'].dt.year
df['Month'] = df['Date'].dt.month
df['Day'] = df['Date'].dt.day
```

```python
df.head()
```

```python
# დავალება 2
```

```python
bins = [0, 5, 7, df['Magnitude'].max()]
labels = ['Low', 'Medium', 'High']
df['Magnitude Category'] = pd.cut(df['Magnitude'], bins=bins, labels=labels, right=False)
```

```python
df.head()
```

```python
# დავალება 3
```

```python
melted_df = pd.melt(df, id_vars=['Magnitude Category'], value_vars=['Depth', 'Magnitude'])
print("Melted DataFrame:")
print(melted_df.head())
```

```python
pivot_df = df.pivot_table(index='Year', columns='Month', values='Magnitude', aggfunc='mean')
print("\nPivot Table:")
print(pivot_df.head())
```

```python
# დავალება 4
```

```python
import matplotlib.pyplot as plt
```

```python
df.plot(x='Year', y='Magnitude', kind='line', title='Magnitude of Earthquakes Over Years')
plt.show()
```

```python
df['Month'].value_counts().sort_index().plot(kind='bar', title='Earthquakes Count by Month')
plt.xlabel('Month')
plt.ylabel('Count')
plt.show()
```

```python
df.plot(x='Depth', y='Magnitude', kind='scatter', title='Magnitude vs Depth of Earthquakes')
plt.show()
```

```python
df['Magnitude'].plot(kind='hist', bins=20, title='Distribution of Earthquake Magnitudes')
plt.xlabel('Magnitude')
plt.ylabel('Frequency')
plt.show()
```
